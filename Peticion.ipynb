```python
import requests

url = "https://api-colombia.com/api/v1/Department"
response = requests.get(url)

if response.status_code == 200:
    data = response.json()
    print(data)
else:
    print(f"Error: {response.status_code}")
```

```python
import pandas as pd
comlombia_df = pd.DataFrame(data)
```

```python
comlombia_df.info
```

```python
comlombia_df.isna().any()
```

```python
df_without_country = comlombia_df.drop(columns="country",axis=1)
df_without_country
```

```python
import matplotlib.pyplot as plt


categories = comlombia_df["name"]
values = comlombia_df["surface"]

plt.barh(categories, values, color='skyblue')
plt.xlabel('Superficie')
plt.ylabel('Departamentos')
plt.title('Superficie por Departamento en Colombia')
plt.show()
```
